```python

from pandas import read_csv
import matplotlib.pyplot as plt 
from matplotlib import pyplot
from pandas.plotting import lag_plot
import seaborn as sns

#import dataset as pandas dataframe
data = read_csv("https://raw.githubusercontent.com/Dobtom/Repo/master/datatraining.txt")
data.head(5)

```

```python
date=data['date']
date.head(5)
```

```python
temp=data['Temperature']
light=data['Light']
carbon=data['CO2']
humidity=data['Humidity']
ratio=data['HumidityRatio']
occu=data['Occupancy']
```

```python
#line plot
plt.plot(date, temp, label='Temperature')
plt.plot(date,light,label='Light')
plt.plot(date,carbon,label='Co2')
plt.plot(date,humidity,label='Humidity')
plt.plot(date,ratio,label='humidity_ratio')
plt.xlabel('date')
plt.legend()
plt.show()
```

```python
#bar chart of temperature and humidity
plt.bar(date,temp,label='Temperature')
plt.bar(date,humidity,label='Humidity')
plt.xlabel('date')
plt.legend()
plt.show()

```

```python
#bar chart of light and Co2
plt.bar(date,light,label='Light')
plt.bar(date,carbon,label='Co2')
plt.legend()
plt.xlabel('date')
plt.show()
```

```python
#histogram of Co2 ,light and humidity
plt.hist(carbon,label='Co2')
plt.hist(light,label='light')
plt.hist(humidity,label='humididty')
plt.xlabel('date')
plt.legend()
plt.show()
```

```python
#density plot of light and Co2
sns.distplot(light,hist=False,kde=True,label='Light')
sns.distplot(carbon,hist=False,kde=True,label='Co2')
plt.legend()
plt.show()
```

```python
#heatmap of Temperature
heat_map=sns.heatmap(temp.to_frame(),cmap="YlGnBu")
plt.ylabel('IDs')
plt.show()
```

```python
#heatmap of Light
heat_map1=sns.heatmap(light.to_frame())
plt.show()
```

```python
#scatter plot of humidity and humidityRatio
plt.scatter(date,humidity,label='humidity')
plt.scatter(date,ratio,label='humidityRatio')
plt.xlabel('date')
plt.ylabel('humidity and humidityRatio')
plt.legend()
plt.show()
```

```python
#boxplot of Temperature
plt.boxplot(temp)
plt.xlabel('temperature')
plt.show()
```

```python
#boxplot of Co2
plt.boxplot(carbon)
plt.xlabel('Co2')
plt.show()
```
