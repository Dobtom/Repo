```python

from pandas import read_csv
import matplotlib.pyplot as plt 
from matplotlib import pyplot
from pandas.plotting import lag_plot
import seaborn as sns

data = read_csv("C:/Users/raksh/Downloads/occupancy_data/datatraining.txt")
data.head(5)

```

```python
date=data['date']
date.head(5)
```

```python
temp=data['Temperature']
light=data['Light']
carbon=data['CO2']
humidity=data['Humidity']
ratio=data['HumidityRatio']
occu=data['Occupancy']
```

```python
plt.plot(date, temp, label='Temperature')
plt.plot(date,light,label='Light')
plt.plot(date,carbon,label='Co2')
plt.plot(date,humidity,label='Humidity')
plt.plot(date,ratio,label='humidity_ratio')
plt.xlabel('date')
plt.legend()
plt.show()
```

```python
plt.bar(occu)
plt.xlabel('Occupancy')
plt.show()

```

```python
plt.bar(date,light,label='Light')
plt.bar(date,carbon,label='Co2')
plt.legend()
plt.xlabel('date')
plt.show()
```

```python
plt.hist(carbon,label='Co2')
plt.hist(light,label='light')
plt.hist(humidity,label='humididty')
plt.legend()
plt.show()
```
